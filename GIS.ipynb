```python
#Import data
import pandas as pd
import numpy as np
df = pd.read_csv('./output.csv')
df
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()

for index, row in df.iterrows():
    if index>6655:
        location = geolocator.geocode(str(row['STATENAME'] + ' '+ row['LCITY'] + ' '+row['LSTREET1']), timeout=100)
        print(index)
        try:
            df["Latitude"][index] = location.latitude
            df["Longitude"][index] = location.longitude
        except:
            df["Latitude"][index] = "ERROR"
            df["Longitude"][index] = "ERROR"
```
